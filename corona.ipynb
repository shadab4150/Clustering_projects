```
from itertools import cycle, islice


def repeatlist(it, count):
    return islice(cycle(it), count)
color = ['r', 'g', 'b','c','m','y','gold','orange','skyblue','pink','cyan','lime','olive','purple']
```

```
import squarify 
```

```
plt.figure(1,figsize=(22,8))
color = list(repeatlist(color, 31))
squarify.plot(sizes=cur[(cur['Confirmed']>100)&(cur['Country/Region']=='China')].Confirmed.values.tolist(),
              label=cur[(cur['Confirmed']>100)&(cur['Country/Region']=='China')]['Province/State'].values.tolist(), alpha=.7,color=color,
              )
plt.axis('off')
plt.title('Number of confirmed cases in CHINA')
plt.show()
```

```
plt.figure(1,figsize=(22,8))
color = list(repeatlist(color, 31))
squarify.plot(sizes=cur[(cur['Confirmed']>15)&(cur['Country/Region']=='US')].Confirmed.values.tolist(),
              label=cur[(cur['Confirmed']>15)&(cur['Country/Region']=='US')]['Province/State'].values.tolist(), alpha=.7,color=color,
              )
plt.axis('off')
plt.title('Number of confirmed cases in USA')
plt.show()
```
